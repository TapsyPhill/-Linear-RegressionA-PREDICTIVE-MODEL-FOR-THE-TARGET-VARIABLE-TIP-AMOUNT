* I need to do the elbow search to find the best k value 

#####    The elblow method to find the best K value for Kmeans 





```python
def bestk(data , max_k):
    mean=[]
    inertias =[]
    
    for k in range(1,max_k):
        kmean= KMeans(n_clusters=k)
        kmean.fit(data)
        
        mean.append(k)
        inertias.append(kmean.inertia_)
        
    sns.scatterplot(x=mean,y=inertias)
    
        
bestk(data=df[['pickup_latitude','pickup_longitude']],max_k=20)        
    
```

Choosing a key value for clustering is not that much of a simple task but however the elbow method does its best to help us find the K value

What it does is it computes the sum of squared  Capital SSE for some values of K for example 123 etc the sum of squared is distances between each member of the cluster and its centroid If you plot the case against the SSE You will see the error decreasing as K gets larger this is because when the number of clusters increases the distance between points and clusters get smaller and so is the error
the idea of the elbow method is to choose the K at which the SSE decreases dramatically making an elbow

###  ElasticNet Regression with a GridSearch


The variables might be useful or useless we don't know in advance end both lesso regression and ridge regression covers one side Either useless variables or useful variables respectively But how do we know when to use lesso regression or Ridge regression If we can't tell the nature of the variables before hand,  this is where we use elastic net regression elastic net regression Combines the regularization in lesso regression and in Ridge regression to make a hybrid, The hybrid elastic net regression is especially good in dealing with situations when they are correlation between parameters but for instance in our data set there is no much correlation between variables but that little correlation will still be dealt with when using ElasticNet Regression, By combining lesso and ridge regression elastic net regression groups and shrinks the parameters associated with the correlated variables And leaves them in equation.

The hyperparameter l1_ratio defines how we mix both Ridge regression and lesso regression regularization. . The acceptable range of values for l1_ratio is: 0 <= l1_ratio <= 1

*  l1_ratio = 0 means there is no  lesso regression term and there is only Ridge regression regularization.
* l1_ratio = 1 means there is no Ridge regression term and there is only lesso regression regularization.
* 0 < l1_ratio < 1 means the regulation is defined as a combination of lesso regression and Ridge regression terms. If l1_ratio is close to 1, it means that the lesso regression term is dominating. If l1_ratio is close to 0, it means that the Ridge regression term is dominating.
* and also the alpha parameter  need to be determined

###  LinearRegression Model
Linear regression is  one of the most important and widely used regression techniques. It’s among the simplest regression methods and One of its main advantages is the ease of interpreting results.When implementing linear regression of some dependent variable y on the set of independent variables X , where the number of predictors,one can assume a linear relationship between y and X. 

The variation of actual responses y, occurs partly due to the dependence on the predictors X. However, there’s also an additional inherent variance of the output.The coefficient of determination, denoted as 𝑅², tells you which amount of variation in y can be explained by the dependence on X , using the particular regression model. A larger 𝑅² indicates a better fit and means that the model can better explain the variation of the output with different inputs.

* it is important to mention Underfitting and Overfitting

One very important question that might arise when you’re implementing polynomial regression is related to the choice of the optimal degree of the polynomial regression function.
There’s no straightforward rule for doing this. It depends on the case. You should, however, be aware of two problems that might follow the choice of the degree: underfitting and overfitting.

###### Underfitting 
occurs when a model can’t accurately capture the dependencies among data, usually as a consequence of its own simplicity. It often yields a low 𝑅² with known data and bad generalization capabilities when applied with new data.

###### Overfitting 
happens when a model learns both data dependencies and random fluctuations. In other words, a model learns the existing data too well. Complex models, which have many features or terms, are often prone to overfitting. When applied to known data, such models usually yield high 𝑅². However, they often don’t generalize well and have significantly lower 𝑅² when used with new data.

### Regression Evaluation Metrics
There are three common evaluation metrics for regression problems


**Mean Absolute Error** (MAE) is the mean of the absolute value of the errors

**Mean Squared Error** (MSE) is the mean of the squared errors

**Root Mean Squared Error** (RMSE) is the square root of the mean of the squared errors


Comparing these metrics:

- **MAE** is the easiest to understand, because it's the average error.
- **MSE** is more popular than MAE, because MSE "punishes" larger errors, which tends to be useful in the real world.
- **RMSE** is even more popular than MSE, because RMSE is interpretable in the "y" units.

All of these are **loss functions**, because we want to minimize them.


```python
from sklearn.metrics  import accuracy_score , mean_absolute_error,mean_squared_error 
print('MAE:', mean_absolute_error(y_test, pridic))
print('MSE:', mean_squared_error(y_test, pridic))
print('RMSE:', np.sqrt(mean_squared_error(y_test, pridic)))
```
